```python
import pandas as pd
from matplotlib import pyplot as plt

import pandas as pd
import re

# Read the CSV file into a DataFrame
tweet_data_int = pd.read_csv('train_nick.csv')
tweet_data_int.head()

# Convert 'tweet' column to string data type and remove non-string entries
tweet_data_int['tweet'] = tweet_data_int['tweet'].astype(str)
tweet_data_int = tweet_data_int[tweet_data_int['tweet'].apply(lambda x: isinstance(x, str))]

# Join the list of words into a single string
tweet_data_int['tweet'] = tweet_data_int['tweet'].apply(lambda x: ' '.join(x))
```

```python
# sample 5% of data to be used as unseen data
data_int = tweet_data_int.sample(frac=0.95, random_state=786)

data_int_unseen = data_int.drop(data_int.index)
data_int.reset_index(inplace=True, drop=True)
data_int_unseen.reset_index(inplace=True, drop=True)

print(data_int_unseen)
# print the revised shape
print('Data for Modeling: ' + str(data_int.shape))
print('Unseen Data For Predictions: ' + str(data_int_unseen.shape))
```
